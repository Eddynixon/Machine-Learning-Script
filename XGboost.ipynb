```
data = pd.read_csv("DataHackTest1.csv") 
```

```
Test = pd.read_csv("Test.csv") 
```

```
Test = Test.drop(['row_id'])
```

```
Data_value = data.drop(['row_id','Ena_derive_1','accepted'],axis=1)
```

```
Data_target = data['accepted']
```

```
X_train, X_test, y_train, y_test = train_test_split(Data_value, Data_target,
                                                    train_size=0.75, test_size=0.25)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```
model = XGBClassifier()
model.fit(X_train, y_train)
```

```

# make predictions for test data
y_pred = model.predict(Test)
predictions = [round(value) for value in y_pred]
```

```
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
